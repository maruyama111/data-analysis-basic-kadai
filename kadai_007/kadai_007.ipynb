```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df
```
