```python
import numpy as np
import matplotlib.pyplot as plt

#折れ線グラフ
y0 = [1,2,-5,2]

#sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

#ヒストグラム
x2 = np.random .normal(15, 5, 2000)

#散布図
np.random.seed(3)
x3 = np.random.normal(0, 2, 60)
y3 = np.random.normal(0, 2, len(x3))

#グラフ描画
fig, axes = plt.subplots(2,3)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))

axes[0,0].plot(y0)
axes[0,1].plot(x1,y1,linewidth=2.0)
axes[0,2].hist(x2)
axes[1,0].scatter(x3,y3)
```
